```python
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize']= (16,9)
plt.style.use("ggplot")
```

```python
#Importing the dataset

data = pd.read_csv('mallCustomerData.txt', sep = ',')
print(data.shape)
data.head(10) #Showing only the first 10 data.
```

```python
print(data['Gender'].value_counts())
```

```python
#Getting the values and plotting it
f1 = data["Annual Income (k$)"].values
f2 = data["Spending Score (1-100)"].values
for key in data.keys():
    print(key)
    
x = np.array(list(zip(f1,f2)))
#x = [[1,1], [1,2],[3,3],[4,4]]
plt.scatter(f1, f2, c = 'blue', s = 20)
plt.title('Visualizing Dataset')
```

```python
from sklearn.cluster import KMeans

#Number of clusters
kmeans = KMeans(n_clusters = 3)
#fitting the input data
kmeans = kmeans.fit(x)
#Getting the clusster labels
labels = kmeans.predict(x)
#centroid values
c = kmeans.cluster_centers_
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x[:,0],x[:,1],x[:,1])
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x[:,0],x[:,1], x[:,1], c = 'y')
ax.scatter(c[:,0],c[:,1], c[:,1], marker = '*', c= '#050505', s= 1000)
```
